# Decision Boundary

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# plotting

fix, ax = plt.subplots(1, 1, figsize=(5,5))

pos = y == 1
neg = y == 0

ax.axis([0,4,0,4])

ax.scatter(X[pos][:,0], X[pos][:,1], marker='x', s = 80, label="y=1", c = 'red')
ax.scatter(X[neg][:,0], X[neg][:,1], marker='o', s = 80, label="y=0", c = 'blue')

ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
ax.set_title('two variable plot')
ax.legend()

plt.show()
```

```python
from _logistic_regression_util import sigmoid
```

```python
# plotting sigmoid function

z_tmp = np.arange(-10,11)
y = sigmoid(z_tmp)

fig, ax = plt.subplots(1,1,figsize=(14,7))

ax.plot(z_tmp, y, c='b')
ax.set_title("Sigmoid function")
ax.set_ylabel("sigmoid(z)")
ax.set_xlabel("z")

ylim = ax.get_ylim()
xlim = ax.get_xlim()

z_threshold = 0
ax.fill_between([xlim[0], z_threshold], [ylim[1], ylim[1]], alpha=0.2, color='blue')
ax.fill_between([z_threshold, xlim[1]], [ylim[1], ylim[1]], alpha=0.2, color='red')

ax.annotate("z >= 0", xy= [z_threshold,0.5], xycoords='data', xytext=[30,5],textcoords='offset points')
ax.annotate("z < 0", xy= [z_threshold,0.5], xycoords='data', xytext=[-50,5],textcoords='offset points', ha='left')
```

### Observations

![](img_lg_reg_obs_1.JPG)

```python
x0 = np.arange(0, 6)
x1 = 3 - x0

fig, ax = plt.subplots(1,1,figsize=(5,5))

ax.plot(x0, x1, c="blue")
ax.axis([0,4,0,4])

ax.fill_between(x0,x1,alpha=0.2)

ax.scatter(X[pos][:,0], X[pos][:,1], marker='x', s = 80, label="y=1", c = 'red')
ax.scatter(X[neg][:,0], X[neg][:,1], marker='o', s = 80, label="y=0", c = 'blue')

ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
ax.set_title('two variable plot')
ax.legend()

plt.show()
```

### Observations

- in the plot above the blue line represents the line x<sub>0</sub> + x<sub>1</sub> - 3 = 0 and it should intersect the x<sub>1</sub> at 3 and x<sub>0</sub> at 3
- the shaded region represents x<sub>0</sub> + x<sub>1</sub> - 3 < 0.
- any point in the shaded region is classified as y = 0, any point above the shaded region is classified as y = 1, this line is know as decision boundary.
- like linear regression, if we use higher order polynomials, we can dray more complex, non-linear decision boundaries.
