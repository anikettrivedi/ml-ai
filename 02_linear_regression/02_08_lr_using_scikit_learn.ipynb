# Linear Regression Using Scikit-Learn Library (SGDRegressor)

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

from _linear_regression_util import load_house_data

np.set_printoptions(precision=2)
```

### Gradient Descent

```python
X_train, y_train = load_house_data()
X_features = ['size(sqft)','bedrooms','floors','age']
```

### Scale/Normalize training data

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)

print (f"peak to peak range without normalization : {np.ptp(X_train, axis=0)}")
print (f"peak to peak range with normalization    : {np.ptp(X_norm, axis=0)}")
```

### Create and fit regression model

```python
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(X_norm, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
```

### View Parameters

```python
b_norm = sgdr.intercept_
w_norm = sgdr.coef_

print(f"model parameters: w: {w_norm}, b: {b_norm}")
```

#### Make predictions

```python
# make prediction using sgdr predict function
y_pred_sgd = sgdr.predict(X_norm)

# make prediction using w, b
y_pred = np.dot(X_norm, w_norm) + b_norm

print(f"\nprediction using np.dot() and sgdr.predict match: {(y_pred == y_pred_sgd).all()}")

print(f"\nPrediction on training set:\n{y_pred[:10]}" )
print(f"\nTarget values \n{y_train[:10]}")
```

### Plotting results

```python
fig, ax = plt.subplots(1,4,figsize=(12,3),sharey=True)

for i in range(len(ax)):
    ax[i].scatter(X_train[:, i], y_train, label="Target", s=15)
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:, i], y_pred, label="Predict", s=14)
    
ax[0].set_ylabel("Price")
ax[0].legend()
fig.suptitle("target vs prediction using z-score normalized model & scikit-learn SGDRegressor")
plt.show()
```
